plotting sin linea
Oara trazar solo marcadores opuedes usar notacion de cadena de acceso directo o que significa "anillos"
Ejemplo:Dibuja dos puntos en el diagrama uno en la posicion (1,3) y oto en la posicion (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,8])
y=np.array([3,10])
plt.plot(x,y,'o')
plt.show()
```

Diagrmas de dispercion(scatter)
se emplean para mostrar la relacion entre 2 variables numericas, detectando correlaciones o agrupamientos, Ejemplo:


```python
import matplotlib.pyplot as plt
import numpy as np
altura=[150,160,170,180,190,200]
peso=[50,60,65,80,85,95]
plt.scatter(altura,peso, color="red",marker='x')
plt.title("relacion entre altura y peso")
plt.xlabel("altura")
plt.ylabel("peso")
plt.show()
```

Graficos de barras (bar y barh)
Sirven para comparar categorias o variables tanto en orientacion vertical como horizontal.
Ejemplo barras verticales:

```python
import matplotlib.pyplot as plt
import numpy as np
categorias=["A","B","C","D"]
Valores=[4,7,1,8]
plt.bar(categorias,Valores, color="orange")
plt.title("grafico de barras")
plt.xlabel("categorias")
plt.ylabel("valores")
plt.show()
```
